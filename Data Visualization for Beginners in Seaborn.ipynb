# Data Visualization for Beginners in Seaborn

In this notebook, we are going to visualize the "Countries of the World" dataset.

# Importing the Libraries

Firstly, we are going to import the necessary libraries for the visualization of the data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Converting the CSV file into Pandas Dataframe
data = pd.read_csv("Countries.csv")
data.head()
```

# Preprocessing the Data for Visualization

In this section, we are going to see the information about the data. Then, we are going to drop the rows having NaN.

```python
data.info()
```

```python
data.isna().sum() #Calculating the number of 'na' values in each rows.
```

```python
data = data.drop(['InfantMortality','NetMigration'],axis=1)
```

```python
data.info()
```

```python
data = data.dropna() #To get the tidy data, we have dropped the rows having 'na' values
data.info()
```

```python
sns.heatmap(data.corr(),cbar=True)# corr() is the function that correlates the quantitative variables
#color bar on the right of heatmap can be enabled by setting true for cbar
plt.show()
```

```python
df=pd.DataFrame()
df['Literacy']=data['Literacy']
df['Phones']=data['Phones']
df['GDP']=data['GDP']
sns.heatmap(df.corr(),annot=True,fmt="f",cbar=False,cmap="YlGnBu",linewidth=1)#defining cmap can change the colours to our desired colours.
#linewidth is the gape between each squares
#setting annot as True will show the correlation values in the heatmap
#fmt changes the format of annotations as integer,float,etc
plt.show()
```

# Customization of Plots

Now, we are going to learn how to customize our plots.

# Adding third variable to categorize the plots

```python
#Adding a third variable in categorical plot
sns.set_style('darkgrid')
hue_colors =sns.color_palette('Blues',5)
sns.countplot(x="Region", data =data, hue='Climate',hue_order=[1,1.5,2,3,4],palette=hue_colors)
plt.xlabel("Regions")
plt.ylabel("Count")
plt.title("The counts of every Region")
plt.xticks(rotation = 270)
plt.show()

sns.color_palette("Paired")
sns.scatterplot(x ='Industry',y ='Service', data = data,hue='GDP')
plt.show()

sns.lineplot(x ='Industry',y ='Service', data = data,hue='GDP')
plt.show()
```

# Setting the background style

Generally there are various styles forsetting backgrounds such as white, dark, etc. Here we can look at a number of styles.

```python
for style in ['white','dark','whitegrid','darkgrid','ticks']:
    sns.set_style(style)
    hue_colors =sns.color_palette('Blues',5)
    sns.countplot(x="Region", data =data, hue='Climate',hue_order=[1,1.5,2,3,4],palette=hue_colors)
    plt.xlabel("Regions")
    plt.ylabel("Count")
    plt.title("The counts of every Region")
    plt.xticks(rotation = 270)
    plt.show()
```

# Modifying the color palettes

```python
sns.set(color_codes=True)#this will set the functionality to use matplotlib color codes
sns.countplot(x=data['Climate'], color='c')#c is for cyan
plt.show()
sns.countplot(x=data['Climate'], color='m')#m is for magenta
plt.show()
```

```python
#this will show the color palettes available in seaborn
for p in sns.palettes.SEABORN_PALETTES:
    sns.set_palette(p)
    sns.palplot(sns.color_palette())
    plt.title(str(p))
    plt.show()
```

# Paired Color Patterns

```python
sns.color_palette("Paired")#for unordered observations
sns.scatterplot(x ='Industry',y ='Service', data = data,hue='GDP')
plt.show()
```

# Sequential Color Patterns

```python
sns.countplot(x='Climate', data=data,order=[4,1.5,1,3,2],palette='Greens')#for ordered observations
plt.show()
```

# Diverging color patterns

```python
sns.countplot(x='Climate', data=data,order=[1,1.5,4,3,2],palette='icefire')#for diverging patterns
#icefire is one of the diverging color pattern available in seaborn
plt.show()
```
